```python
import numpy as np
import matplotlib.pyplot as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False #minus 표시
mpl.rc('font', family='Malgun Gothic')  # 폰트 설정(맑은 고딕)
%matplotlib inline

import csv

f = open('doctor2.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

list_data = []
for line in lines :
    list_data.append(line[:])




region = []
doctor_nomal = []
doctor_intern = []
doctor_res = []
doctor_pro = []
k=2

for i in range(k,len(list_data),2) :
    region.append(list_data[i][0])
    doctor_nomal.append(float(list_data[i][1]))
    doctor_intern.append(float(list_data[i][2]))
    doctor_res.append(float(list_data[i][3]))
    doctor_pro.append(float(list_data[i][4]))

x = region
y1 = doctor_nomal
y2 = doctor_intern
y3 = doctor_res
y4 = doctor_pro


plt.plot(x, y1, label='일반의', color='red', linestyle='-', marker='o')
plt.plot(x, y2, label='인턴', color='blue', linestyle='--', marker='s')
plt.plot(x, y3, label='레지던트', color='yellow', linestyle=':', marker='d')
plt.plot(x, y4, label='전문의', color='magenta', linestyle='-.', marker='o')

plt.ylim(0, 10000, 2000)
plt.title('서울 및 광역시 종합병원 의사 수')
plt.legend(loc='best')

plt.show()
```
