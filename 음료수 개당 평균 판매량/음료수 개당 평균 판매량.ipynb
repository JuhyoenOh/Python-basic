```python
a=[]
b=[]
for i in range(len(f)):
    k=f.loc[i]
    
    if k[1]=='스포츠,이온음료' :
        a.append(k)
    elif k[1]=='에너지음료':
        b.append(k)
```

```python
a=pd.DataFrame(a)
```

```python
b=pd.DataFrame(b)
```

```python
a_YM=a.loc[:,['YM']]
a_QTY=a.loc[:,['QTY']]


b_YM=b.loc[:,['YM']]
b_QTY=b.loc[:,['QTY']]
```

```python
import numpy as np
import matplotlib.pyplot as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False #minus 표시
mpl.rc('font', family='Malgun Gothic')  # 폰트 설정(맑은 고딕)
%matplotlib inline
```

```python
#각 월 판매량

xdata1 = a_YM
xdata2 = b_YM
ydata1 = a_QTY
ydata2 = b_QTY

fig, axs = plt.subplots(nrows =1 , ncols=2, figsize=(9,5), sharex=True, sharey=True)
ax = axs[0]
ax.plot(xdata1,ydata1, color='red', linestyle='--',marker='o')
ax.set_title('스포츠,이온음료')

ax = axs[1]
ax.plot(xdata2, ydata2, linestyle='--',marker='o')
ax.set_title('에너지음료')


plt.subplots_adjust(left = 0.1, right=0.9, bottom=0.1, top=1, wspace=0.2, hspace=0.3)



plt.show()
```
