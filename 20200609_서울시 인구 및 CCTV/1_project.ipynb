# 서울시 CCTV 현황 데이터 분석

```python
import numpy as np
import pandas as pd
```

```python
pop= pd.read_csv('서울시_인구현황.txt',thousands=',', skiprows=[0,1,2], sep='\t', 
                 usecols=[1,3,6,9,13],names = ['자치구', '인구수', '내국인', '외국인', '고령자' ])
pop
```

```python
#외국인 비율 = 서울(구) 외국인 인구/ 서울(구) 총 인구
#고령자 비율 = 서울(구) 고령자 인구/ 서울(구) 총 인구

pop['외국인 비율'] = (pop['외국인']/pop['인구수'])*100
pop['고령자 비율'] = (pop['고령자']/pop['인구수'])*100
pop
```

```python
cctv= pd.read_excel('서울시 자치구 년도별 CCTV 설치 현황.xlsx',thousands=' ')

cctv
```

```python
cctv['기관명'].unique()
```

```python
#cctv['기관명']= ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구',
#       '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구',
#       '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구',
#       '은평구', '종로구', '중구', '중랑구']


cctv['기관명']= [x.replace(' ', '')for x in cctv['기관명'].unique()]
```

```python
cctv.rename(columns={'기관명':'자치구'}, inplace=True)  ## inplace=True 원본 데이터를 바꿔라.
cctv.head()
```

```python
cctv[['2011년 이전','2012년','2013년']]=\
    cctv[['2011년 이전','2012년','2013년']].fillna(0).astype(int)
cctv
```

```python
cctv['2015년 이전'] = (cctv['2011년 이전']+cctv['2012년']+
                    cctv['2013년']+cctv['2014년']+cctv['2015년'])
cctv['최근 3개년'] = (cctv['2016년']+cctv['2017년']+cctv['2018년'])
## 
cctv
```

```python
cctv1=cctv.iloc[:,[0,1,10,11]]
cctv1
```

```python
cctv1['최근 증가율'] = (cctv1['최근 3개년']/cctv1['2015년 이전'])*100
cctv2=cctv1.iloc[:,[0,1,4]]
cctv2

# del.cctv1['2015년 이전']   >>> 이렇게 하면 그 부분의 테이블만 삭제가 된다.
```

```python
result=pd.merge(pop,cctv2,left_on='자치구',right_on='자치구',how='inner')
result
```

```python
result.set_index('자치구'),inplace=True

result
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False #minus 표시
mpl.rc('font', family='Malgun Gothic')  # 폰트 설정(맑은 고딕)
%matplotlib inline

plt.rcParams['figure.figsize']=(12,6)
plt.rcParams['font.size']=14


result1=result.iloc[:,[0]]
result2=result.iloc[:,[7]]

x= result2
y= result1
plt.bar(x,y)
plt.title("구별 CCTV 개수")
plt.xlabel("CCTV 개수")
plt.ylabel("자치구")
plt.show
```

### 자치구별 최근 증가율 (내림차순으로)

```python
df = result.set_index('자치구')

df['소계'].sort_values().plot(kind='barh', grid=True, figsize=(12,10))

plt.show()
```

   ## 인구수와 CCTV 댓수의 산점도

```python
plt.figure(figsize=(12,10))
plt.scatter(result['인구수'],result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('cctv 대수')
plt.grid()
plt.show()
```

  ## 인구수와 CCTV 댓수의 산점도 및 회귀선

```python
fp1 = np.polyfit(result['인구수'],result['소계'],1)
fp1
```

```python
f1 =  np.poly1d(fp1)
fx = np.linspace(100000,700000,101)
plt.figure(figsize=(12,10))

plt.scatter(result['인구수'],result['소계'], s=50)
plt.plot(fx,f1(fx), ls='-.',lw=3,color='red')
plt.xlabel('인구수')
plt.ylabel('cctv 대수')
plt.grid()
plt.show()
```

##### 인구수와 CCTV댓수의 산점도,회귀선 및 자치구명

```python
df.head()
```

```python
df['오차']=np.abs(df['소계']-f1(df['인구수']))

df.head()
```

```python
df_sort = df.sort_values(by=['오차'], ascending = False)
df_sort.head(10)
```

```python


plt.figure(figsize=(14,10))
plt.scatter(df['인구수'],df['소계'],c=df['오차'],s=50)
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='g')

for n in range(10) :
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize= 15)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
